## Importaci√≥n

```python
import seaborn as sns
```

## Lectura de dataset de la biblioteca de *Seaborn*

```python
df = sns.load_dataset('tips')
df.head(5)
```

## Countplot

#### Sobre una variable categ√≥rica

```python
sns.countplot(x ='sex', data = df);
```

#### Sobre dos variables categ√≥ricas

```python
sns.countplot(x ='sex', hue = 'smoker', data = df); # cambiando el par√°metro x por y, ser√≠a horizontal
```

## Displot
`Distplot` est√° desaconsejado, y se recomienda utilizar `Displot` en su lugar.

```python
sns.displot(data=df, x="total_bill").fig.set_size_inches(12,6);
```

## Boxplot

```python
sns.boxplot(x='day', y='tip', data=df);
```

Se puede obtener la informaci√≥n de los diagramas de cajas en un diccionario. De esta forma, se puede acceder, entre otros, a los valores at√≠picos (*outliers*).

```python
from matplotlib.cbook import boxplot_stats  

propJueves = df[df['day']=='Thur']['tip'].values
diccioStats=boxplot_stats(propJueves)
diccioStats
```

## Barplot
Proporciona una medida de tendencia central (promedio t√≠picamente) junto con el intervalo del confianza del 95%.

```python
sns.barplot(x = 'day', y = 'total_bill', data = df);
```

## Pairplot
Muestra m√∫ltiples distribuciones bivariadas de conjuntos de datos

```python
sns.pairplot(df);
```

#### Dividiendo con una categor√≠a

```python
sns.pairplot(df, hue ='sex');
```

## Mapas de calor (*heatmap*)
Muestran la informaci√≥n en forma de colores para cada par de valores en los ejes x e y.

```python
import pandas as pd
df = pd.pivot_table(data=sns.load_dataset("flights"),
                    index='month',
                    values='passengers',
                    columns='year')
df.head()
```

```python
sns.heatmap(df);
```

Los mapas de calor se usan frecuentemente para mostrar la correlaci√≥n entre distintas columnas de un dataframe de Pandas. 

```python
sns.heatmap(df.corr());
```
