### Gráfica única

```
plt.figure(figsize=(8,6))                     # para pérdidas:
plt.plot(historial.history['accuracy'])       #    'mse'
plt.plot(historial.history['val_accuracy'])   #    'val_mse'
plt.title('Exactitud (accuracy)')             #    'Error cuadrático medio'
plt.ylabel('Exactitud (accuracy)')            #    'MSE'
plt.xlabel('Epoch');
plt.legend(['Entrenamiento', 'Validación'], loc='upper left') # 'upper right', 'best'
```


### Gráficas en grid

```
fig, axs = plt.subplots(2, figsize=(8,10))
fig.suptitle('Dropout y batchnorm')

axs[0].plot(historial.history['loss'])
axs[0].plot(historial.history['val_loss'])
axs[0].set_title('Categorical Crossentropy Loss')

axs[0].set(ylabel='loss', xlabel='epoch')
axs[0].legend(['Entrenamiento', 'Validación'], loc='upper right')

axs[1].plot(historial.history['accuracy'])
axs[1].plot(historial.history['val_accuracy'])
axs[1].legend(['Entrenamiento', 'Validación'], loc='lower right')
axs[1].set_title('Accuracy')

axs[1].set(xlabel='Epoch', ylabel='Accuracy');
```
